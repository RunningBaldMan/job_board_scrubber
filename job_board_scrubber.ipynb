```python
def get_remoteok_jobs(tags=None):
    url = os.getenv('REMOTEOK_URL')
    if tags:
        url += f"?tags={','.join(tags)}"

    headers={"User-Agent": "Mozilla/5.0"} # Required for access
    response = requests.get(url, headers=headers)

# Handle errors

    if response.status_code == 200:
        jobs = response.json()
        print(f"Successfully retrieved {len(jobs)} jobs listings!")
        return jobs
    else:
        print(f"Failed to retrieve data. Status Code: {response.status_code}")
        return[]
```
